# Step 6: Building Dynamic Price Hotel bocking Model 

```python
# =========================
# Step 1: Cancellation rate by month
# =========================

cancel_by_month = df.groupby("arrival_date_month")["is_canceled"].mean()

df["cancel_rate_month"] = df["arrival_date_month"].map(cancel_by_month).astype(float)


# =========================
# Step 2: Create Season Column
# =========================

def get_season(month):
    if month in ["December", "January", "February"]:
        return "Low"
    elif month in ["March", "April", "May", "September", "October", "November"]:
        return "Mid"
    else:
        return "High"

df["season"] = df["arrival_date_month"].apply(get_season)


# =========================
# Step 3: Assign Season Factor
# =========================

season_factor_map = {
    "Low": -0.10,   # تخفيض
    "Mid": 0.05,    # زيادة بسيطة
    "High": 0.20    # زيادة قوية
}

df["season_factor"] = df["season"].map(season_factor_map)


# =========================
# Step 4: Dynamic Pricing (Month + Season)
# =========================

df["dynamic_price"] = df["adr"] * (
    1
    - 0.4 * df["cancel_rate_month"]   # تأثير الإلغاء
    + df["season_factor"]             # تأثير الموسم
)


# =========================
# Step 5: Compare Results
# =========================

print(df[["arrival_date_month", "adr", "cancel_rate_month", "season", "dynamic_price"]].head())
```

```python
# =========================
# 1. Average comparison
# =========================

avg_adr = df["adr"].mean()
avg_dynamic = df["dynamic_price"].mean()

plt.figure()
plt.bar(["Original ADR", "Dynamic Price"], [avg_adr, avg_dynamic])
plt.title("Average Price Before vs After")
plt.ylabel("Price")
plt.show()


# =========================
# 2. Monthly trend
# =========================

monthly = df.groupby("arrival_date_month")[["adr", "dynamic_price"]].mean()

monthly.plot()
plt.title("Monthly Price Trend")
plt.ylabel("Price")
plt.show()


# =========================
# 3. Season distribution
# =========================

season_counts = df["season"].value_counts()

plt.figure()
plt.pie(season_counts, labels=season_counts.index, autopct='%1.1f%%')
plt.title("Season Distribution")
plt.show()
```

# Step 7:(Revenue Optimization)
The implementation of a dynamic pricing strategy improved overall revenue by aligning prices 
with demand patterns and seasonal trends.

```python
# =========================
# Step 1: Revenue Calculation
# =========================

# الإيراد قبل (نفترض كل حجز = ليلة وحدة)
df["revenue_before"] = df["adr"]

# الإيراد بعد
df["revenue_after"] = df["dynamic_price"]


# =========================
# Step 2: Total Revenue
# =========================

total_before = df["revenue_before"].sum()
total_after = df["revenue_after"].sum()

print("Revenue Before:", total_before)
print("Revenue After:", total_after)


# =========================
# Step 3: Revenue by Month
# =========================

monthly_revenue = df.groupby("arrival_date_month")[["revenue_before", "revenue_after"]].sum()

monthly_revenue.plot(title="Revenue Before vs After by Month")
```
